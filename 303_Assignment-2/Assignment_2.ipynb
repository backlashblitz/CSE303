#1. Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
```

Convert Categorical Data to Numerical using Label Encoding

```python
label_encoder = {}
for column in df.columns:
    if df[column].dtype == 'object':
        le = LabelEncoder()
        df[column] = le.fit_transform(df[column])
        label_encoder[column] = le
```

Normalize numerical columns

```python
# Identify numerical columns, excluding the target variable 'Risk_Level'
numerical_cols = df.select_dtypes(include=['number']).columns.drop('Risk_level', errors='ignore')
scaler = MinMaxScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
df
```

```python
# Split the data into training and testing sets
X = df.drop('Risk_level', axis=1)
y = df['Risk_level']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) # 70% training and 30% test
```

#3. Model Training and Evaluation

1.Random Forest

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score
```

```python
# Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
```

2.XGBoost

```python
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')

xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_test)
```

3.Logistic Regression

```python
# Logistic Regression
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
y_pred_lr = lr_model.predict(X_test)
```

#Evaluate models using appropriate metrics:

```python
# Evaluation function to print the metrics
def evaluate_model(y_test, y_pred, model_name, model):
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='macro', zero_division=0)
    recall = recall_score(y_test, y_pred, average='macro', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='macro', zero_division=0)
    # Added multi_class='ovr' to handle multi-class problems,
    # use predict_proba to get the probability of each class
    y_pred_proba = model.predict_proba(X_test)
    roc_auc = roc_auc_score(y_test, y_pred_proba, average='macro', multi_class='ovr')

    print(f"Model: {model_name}")
    print(confusion_matrix(y_test, y_pred))
    print(f"{model_name} Performance:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print(f"ROC-AUC: {roc_auc:.4f}\n")


# Evaluating all models
evaluate_model(y_test, y_pred_rf, 'Random Forest', rf_model)
evaluate_model(y_test, y_pred_xgb, 'XGBoost', xgb_model)
evaluate_model(y_test, y_pred_lr, 'Logistic Regression', lr_model)
```

#Model performance comparison table






```python
models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss')
}

results = [] # to store all the results
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='macro', zero_division=0)
    recall = recall_score(y_test, y_pred, average='macro', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='macro', zero_division=0)
    y_pred_proba = model.predict_proba(X_test)
    roc_auc = roc_auc_score(y_test, y_pred_proba, average='macro', multi_class='ovr')

    results.append({
        'Model': model_name,
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'ROC-AUC': roc_auc
    })
df_performance = pd.DataFrame(results)
df_performance
```

#ROC Curves for all Models

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
import numpy as np
import matplotlib.pyplot as plt

# Binarize the output
y_test_bin = label_binarize(y_test, classes=np.unique(y))
n_classes = y_test_bin.shape[1]

# Set up the plot
plt.figure(figsize=(10, 8))

# Initialize a counter for the models
model_counter = 0

# Loop through each model
for model_name, model in models.items():
    y_pred_proba = model.predict_proba(X_test)

    # Compute micro-average ROC curve and ROC area
    fpr, tpr, _ = roc_curve(y_test_bin.ravel(), y_pred_proba.ravel())
    roc_auc = auc(fpr, tpr)

    # Plot ROC curve for the current model
    plt.plot(fpr, tpr,
             label=f'{model_name} (AUC = {roc_auc:.2f})')

    # Increment the model counter
    model_counter += 1
plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')

# Set labels and title
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for All Models')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

#4.Hyperparameter Tuning

Selecting the best-performing model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Example models
models = {
    "Random Forest": RandomForestClassifier(random_state=42),
         "Logistic Regression": LogisticRegression(random_state=42, max_iter=1000),
         "XGBoost": XGBClassifier(random_state=42)
}

# Evaluate each model
model_scores = {}
for name, model in models.items():
    model.fit(X_train, y_train)  # Ensure X_train and y_train are defined
    y_pred = model.predict(X_test)  # Ensure X_test is defined
    accuracy = accuracy_score(y_test, y_pred)  # Ensure y_test is defined
    model_scores[name] = accuracy

# Print model scores
print("Model Scores:", model_scores)

# Select the best-performing model
best_model_name = max(model_scores, key=model_scores.get)
print(f"Best-Performing Model: {best_model_name}")
```

Tuning its hyperparameters using Random search

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV

# Reduced parameter grid
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# Random Forest model
model = RandomForestClassifier(random_state=42)

# Randomized Search with fewer combinations
randomized_search = RandomizedSearchCV(
    estimator=model, param_distributions=param_grid,
    n_iter=10, scoring='accuracy', cv=3, n_jobs=-1, random_state=42, verbose=1
)
randomized_search.fit(X_train, y_train)

# Results
print("Best Parameters:", randomized_search.best_params_)
print("Best Cross-Validation Score:", randomized_search.best_score_)
test_score = randomized_search.best_estimator_.score(X_test, y_test)
print("Test Score:", test_score)
```

#5. Feature Importance Analysis

Use the best-performing model to analyze feature importance and Bar chart for feature importance.

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
import numpy as np

# Assume X_train and y_train are already defined (X_train is a DataFrame)
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Get feature importances
importances = model.feature_importances_

# Create an array of feature names (assuming X_train is a pandas DataFrame)
feature_names = X_train.columns if hasattr(X_train, 'columns') else np.array([f"Feature {i+1}" for i in range(X_train.shape[1])])

# Sort the feature importances in descending order
indices = np.argsort(importances)[::-1]

# Number of top features you want to display
top_n = 10  # Change this number to display more than 5 features

# Get the top N features and their importance scores
top_n_features = feature_names[indices[:top_n]]
top_n_importances = importances[indices[:top_n]]

# Plot the top N feature importances
plt.figure(figsize=(12, 6))
plt.title(f"Top {top_n} Important Feature")
plt.bar(range(top_n), top_n_importances, align="center")
plt.xticks(range(top_n), top_n_features, rotation=90)
plt.xlabel("Feature")
plt.ylabel("Importance")
plt.show()
```
