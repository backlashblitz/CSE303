#2. Feature Engineering

```python
df['Study_Hours_per_Week'].max()
#S.N: Maximum Study_Hours_per_Week from the dataset is 39
```

```python
df['Attendance_Percentage'].max()
#S.N: Maximum Attendance_Percentage from the dataset is 99.99
```

```python
def mod_Study_Hours_per_Week(x):
  return(x/40)*50

def mod_Attendance_Percentage(x):
  return(x/100)*50;

df['Engagement_Score']= df['Study_Hours_per_Week'].apply(mod_Study_Hours_per_Week) + df['Attendance_Percentage'].apply(mod_Attendance_Percentage)

 #S.N-1: A new feature named 'Engagement_Score' has been created by combining 'Study_Hours_per_Week' and 'Attendance_Percentage'.
 #S.N-2: The 'Study_Hours_per_Week' metric has been assigned a weight of 50%, relative to a standard study time of 40 hours per week.
 #S.N-3: The 'Attendance_Percentage' metric has been assigned a weight of 50%, based on the total attendance being considered as 100%.
```

```python
df
```

```python
def mod_Academic_Standing(x):
  if x >= 3.50:
    return 'High'
  elif x >= 3.00:
    return 'Average'
  else:
    return 'Low'

df['Academic_Standing'] = df['CGPA'].apply(mod_Academic_Standing)

 #S.N : A feature named Academic Standing that categorizes CGPA into ’Low’, ’Average’, ’High’ based on thresholds.
```

```python
df
```

```python

def quantify_Satisfaction_Level(x):
  if x == 'Very Satisfied':
    return 50
  elif x == 'Satisfied':
    return 40
  elif x == 'Neutral':
    return 30
  elif x == 'Dissatisfied':
    return 10
  elif x == 'Very Dissatisfied':
    return 0

df['qun_Satisfaction_Level'] = df['Satisfaction_Level'].apply(quantify_Satisfaction_Level)



def quantify_Part_Time_Job(x):
  if x == 'Yes':
    return 10
  elif x == 'No':
    return 0

df['qun_Part_Time_Job'] = df['Part_Time_Job'].apply(quantify_Part_Time_Job)

#S.N-1: Quantify the Categorical datas to Numerical datas for 'Satisfaction_Level' and 'Part_Time_Job'.
#S.N-2: A new column, qun_Satisfaction_Level, is created in the DataFrame to store these numerical values.
#S.N-3: The quantify_Part_Time_Job function changes "Yes" and "No" in the 'Part_Time_Job' column into numerical values.
```

```python
df
```

```python
df['qun_Risk_level'] = df['qun_Satisfaction_Level'] + df['Attendance_Percentage']*0.5 + df['qun_Part_Time_Job']

def mod_Risk_level(x):
  if x >= 70:
    return 'Low'
  elif x >= 40:
    return 'Neutral'
  else:
    return 'High'

df['Risk_level'] = df['qun_Risk_level'].apply(mod_Risk_level)

#S.N: The Attendance_Percentage metric has been assigned a weight of 50%, contributing proportionally to the overall Risk_level calculation.
      #This, along with the quantified Satisfaction_Level and Part_Time_Job metrics, determines the qun_Risk_level.
```

# 3.Outlier Detection


```python
q1 = df['Age'].quantile(0.25)
q3 = df['Age'].quantile(0.75)
IQR = q3 - q1


lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR


outlier_count = 0

for i in df['Age']:
    if i < lower_bound or i > upper_bound:
        outlier_count += 1

print("Number of outliers:", outlier_count)

#S.N: Using IQR we calculate the number of Outliers.There are no outliers in the 'Age' column.
```

#4. Advanced Data Manipulation

```python
pivot_table = df.pivot_table(
    values='CGPA',
    index='Department',
    columns='Year_of_Study',
    aggfunc='mean'
)

print("Pivot Table: Average CGPA by Department and Year of Study")
print(pivot_table)

#S.N: A pivot table to summarize average CGPA by Department and Year of Study.
```

```python
def quantify_Family_Income(x):
    if x == 'Low':
        return 10
    elif x == 'Medium':
        return 20
    elif x == 'High':
        return 30

# Applying the function to the Family_Income column
df['qun_Family_Income'] = df['Family_Income'].apply(quantify_Family_Income)

#S.N: Quantify the Categorical datas to Numerical datas for 'Family_Income'.
```

```python
df
```

```python
def is_high_risk(row):

    if row['Attendance_Percentage'] < 50 and row['qun_Part_Time_Job'] == 0 and row['qun_Family_Income'] <= 10:
        return 'High Risk'
    else:
        return 'Not High Risk'


df['High_Risk_Status'] = df.apply(is_high_risk, axis=1)


df

#S.N: a new feature that indicates whether a student is at high risk based on multiple factors (e.g., Attendance, Part-Time Job, and Family Income).
```

```python

engagement_grouped = df.groupby(['Gender', 'Department']).agg({'Engagement_Score':'mean'})


print("Engagement Score Trends by Gender and Department:")
print(engagement_grouped)

#S.N-1: Calculates the average Engagement_Score for each combination of Gender and Department.
#S.N-2: The mean of the 'Engagement_Score' is computed for each group using the .agg() method.
```
